```python
aqi['Date'] = pd.to_datetime(aqi['Date'], format='%B-%Y')
aqi
```

```python
aqi['Month'] = aqi['Date'].dt.month
aqi['Year'] = aqi['Date'].dt.year
```

```python
aqi
```

```python
aqi = aqi[['City', 'Date', 'Month', 'Year', 'PM2.5', 'PM10', 'NO', 'NO2', 'NH3', 'CO', 'SO2', 'O3', 'AQI', 'AQI_Bucket']]
aqi
```

```python
conditions = [
    aqi['Month'].isin([12, 1, 2]),
    aqi['Month'].isin([3, 4, 5]),
    aqi['Month'].isin([6, 7, 8, 9]),
    aqi['Month'].isin([10, 11])
]

choices = ['Winter', 'Summer', 'Monsoon', 'Post-Monsoon']

aqi['Season'] = np.select(conditions, choices, default='unknown')
aqi
```

```python
aqi = aqi[['City', 'Date', 'Month', 'Year', 'Season', 'PM2.5', 'PM10', 'NO', 'NO2', 'NH3', 'CO', 'SO2', 'O3', 'AQI', 'AQI_Bucket']]
```

```python
aqi
```

```python
aqi['PM2.5'] = aqi['PM2.5'].fillna(
    aqi.groupby(['City', 'Month'])['PM2.5'].transform('mean')
)

aqi.head(100)
```

```python
columns_to_fill = ['PM10', 'NO', 'NO2', 'NH3', 'CO', 'SO2', 'O3']

for col in columns_to_fill:
    aqi[col] = aqi[col].fillna(
        aqi.groupby(['City', 'Month'])[col].transform('mean')
    )
```
